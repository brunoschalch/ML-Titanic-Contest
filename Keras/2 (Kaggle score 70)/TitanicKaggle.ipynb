```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
validate = pd.read_csv('gender_submission.csv')
```

# Drop features-... is this good?

```python
train = train.drop(['Name','SibSp', 'Ticket', 'Cabin', 'Embarked'],axis=1)
test = test.drop(['Name','SibSp', 'Ticket', 'Cabin', 'Embarked'],axis=1)

train
```

# Sex mapping 1 - male, 0 - female

```python
train['Sex_binary']=train['Sex'].map({'male':1,'female':0})
test['Sex_binary']=test['Sex'].map({'male':1,'female':0})
```

# Age cleaning

```python
train['Age'] = train['Age'].fillna(29)
test['Age'] = test['Age'].fillna(29)
```

```python
features = ['Pclass','Age','Sex_binary', 'Fare', 'Parch']
target = 'Survived'
```

```python
train[features].head(3)
```

```python
import numpy as np
```

```python
X_train = np.array(train[features])
X_train
```

```python
y_train = np.array(train[target])
y_train = y_train.reshape(-1,1)
```

```python
# Import keras modules
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import glorot_normal
from keras.optimizers import SGD
```

```python
model = Sequential()
numNeurons = 4
model.add(Dense(numNeurons, input_dim=len(features), activation='relu', kernel_initializer="uniform"))
model.add(Dense(1, activation='sigmoid', kernel_initializer="uniform"))
#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['mae']) old

sgd = SGD(lr=0.08, decay=1e-6, momentum=0.7, nesterov=True)

model.compile(loss='mean_squared_logarithmic_error', optimizer=sgd, metrics=['mae'])
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=20)
#score = model.evaluate(X_test, y_test, batch_size=25)
#score
```

```python
X_test = np.array(test[features])
X_test
```

```python
predicciones = model.predict(X_test)
#print(predicciones)
```

```python
predicciones = predicciones.tolist()
```

```python
pre = pd.Series(predicciones)
validate['prediccion'] = pre
validate['prediccion'] = validate['prediccion'].str.get(0)
validate
```

```python
coincidencias = []
for dato in validate.prediccion:
    if dato >= 0.5:
        coincidencias.append(1)
    else:
        coincidencias.append(0)
validate['final'] = coincidencias
validate
```

```python
coincide = 0
coincide = sum(validate['Survived'] == validate['final'])
print(coincide)
print(float(coincide)/ float(len(validate)))


match = 0
nomatch = 0
for val in validate.values:
    if val[1] == val[3]:
        match = match +1
    else:
        nomatch = nomatch +1
print(float(match)/float(len(validate)))
```

```python
toKaggle = pd.DataFrame({'PassengerId':validate['PassengerId'],
                         'Survived':validate['final']})

toKaggle.head()
```
